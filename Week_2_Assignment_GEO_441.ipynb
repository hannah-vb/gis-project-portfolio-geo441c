## **Reflection on data management function in ArcGIS Pro**

- **Write a 100-word paragraph to describe the relationship between**
  - **`Geodatabase`**
  - **`Feature Dataset`**
  - **`Feature Classes`**
  - I can refer to **Page 9** of [assignment guide](https://miamioh.instructure.com/courses/246872/files/folder/week-02-resources?preview=38195670) or other resources online, such as
    - [Fundamentals of geodatabase](https://pro.arcgis.com/en/pro-app/latest/help/data/geodatabases/overview/fundamentals-of-the-geodatabase.htm)
    - [Feature datasets in ArcGIS Pro](https://pro.arcgis.com/en/pro-app/latest/help/data/feature-datasets/feature-datasets-in-arcgis-pro.htm)

A Geodatabase is a place to store all different kinds of data in one section of the file explorer. Raster, vector, points, polygons, and topology can all be stored together when they are stored in geodatabases. Geodatabases contain Feature Datasets which are collections of spatial data. Within this spatial data, there are feature classes that make up the feature datasets. Examples of feature classes include locations of rivers, locations of lakes, and locations of vegetation. Then, when all of the feature classes are put together that makes a feature dataset. The geodatabase is the overall container, the feature dataset is held within the geodatabase, and the feature class is a layer of data that is held within the feature dataset.
